```python
import osmnx as ox
```

```python
city_query = 'Astana, Kazakhstan'
gdf = ox.geocode_to_gdf(city_query)
```

```python
gdf_bound = gdf.total_bounds
e,s,w,n = gdf_bound
(n,s,w,e)
```

```python
fp = gdf.geometry[0]
fp
```

```python
fig, ax = ox.plot_footprints(gdf, figsize=(30, 30), color='none', edge_color='#D9381E', edge_linewidth=1, alpha=None, 
                             bgcolor='none', bbox=(n,s,w,e), 
                             save=True, show=True, close=False, filepath='astana_bond.png', dpi=600)
```

```python
G = ox.graph_from_place('Astana, Kazakhstan', network_type='all', simplify=True, retain_all=True)
```

```python
fig2, ax2 = ox.plot_graph(G, figsize=(30, 30), bgcolor='#BCB88A', edge_color='#CCCCFF',
                        node_size=0, edge_linewidth=0.3,
                        show=True, close=False, save=False, filepath='astana_coloured.png', 
                          dpi=300, bbox=(n,s,w,e))
```
